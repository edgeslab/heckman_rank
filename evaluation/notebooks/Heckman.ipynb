```python
def gen_dummy_data(train=True, N=1000, DX=100, DZ=2, sigma=0.5):
    df = pd.DataFrame()
    
    df['qid'] = pd.Series(np.random.choice(int(N/20), N))
    df['C'] = pd.Series(np.random.choice(2, N))

    for col in ['X'+str(n) for n in range(DX)]:
        df[col] = pd.Series(np.random.rand(N))

    if train: 
        df['S'] = pd.Series(np.random.choice(2, N))
        df['eps'] = pd.Series(np.random.normal(0.0, sigma, N))
        for col in ['Z'+str(n) for n in range(DZ)]:
            df[col] = pd.Series(np.random.rand(N))
    
    df = df.sort_values(by=['qid'])
    
    return df
    
```

### First Step: Probit

```python
def probit(Y, X):
#     pm = sm.Probit(endog=Y, exog=X)
#     return pm.fit(disp=0).params
    lr = LinearRegression()
    lr.fit(X, Y)
    return lr.coef_
```

```python
X = train[train.columns.drop(['qid', 'C', 'S'])]
Y = train['S']

N = X.shape[0]
# v = np.random.normal(0.0, 1.0, N)

# gamma = probit(Y, np.append(X.values, v.reshape(-1,1), 1))[:-1]
gamma = probit(Y, X)
print(gamma.shape)
gamma[:5]
```

### Question: should gamma be normalized?

```python
gamma.min(), gamma.max(), gamma.sum()
```

### Calculate lambda

```python
def inverse_mills(val):
    return norm.pdf(val) / (1 - norm.cdf(val))

lambda_ = inverse_mills(np.matmul(X, gamma))
lambda_.shape
```

### Fit Model Parameter

```python
def OLS(Y, X):
    N = X.shape[0]
    lr = LinearRegression()
    lr.fit(X, Y)
    return lr.coef_
```

```python
xcols = list(filter(lambda x: x[0] == 'X', train.columns))
X = train[xcols]
Y = train['C']

X.values.shape
lambda_.reshape(-1,1).shape
np.append(X, lambda_.reshape(-1,1), 1).shape

params = OLS(Y, np.append(X, lambda_.reshape(-1,1), 1))[:-1]
params.shape
```

### Inference

```python
X_test = test[test.columns.drop(['qid', 'C'])]
Eval = test[['qid', 'C']]
Eval['Relv'] = np.matmul(X_test, params)
Eval.head(10)
```

### Evaluation

```python
avg_rank = 0.0
cnt = 0
for name, group in Eval.groupby(['qid']):
    df = group
    df['ranks'] = df['Relv'].rank(ascending=False).astype(int)
    avg_rank += df[df['C'] == 1]['ranks'].mean()
    cnt += 1
print('average rank of clicked docs: %0.4lf' % (avg_rank / cnt))
```
